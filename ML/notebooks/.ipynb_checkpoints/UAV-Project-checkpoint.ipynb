```python
#Importing the necessary library's 
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#Reading experimental csv files from Dataset folder
df1 = pd.read_csv("../Dataset/experiment1.csv")
df2 = pd.read_csv("../Dataset/experiment2.csv")
df3 = pd.read_csv("../Dataset/experiment3.csv")
#Appending all the experimental csv files to single dataframe ie:experimental_df
experimental_df = pd.concat([df1, df2, df3], ignore_index=True)
#Reading ggeometrical csv files from Dataset folder
df1 = pd.read_csv("../Dataset/geometric1.csv")
df2 = pd.read_csv("../Dataset/geometric2.csv")
df3 = pd.read_csv("../Dataset/geometric3.csv")
#Appending all the ggeometrical csv files to single dataframe ie:geometrical_df
geometrical_df = pd.concat([df1, df2, df3], ignore_index=True)
experimental_df.columns=experimental_df.columns.str.lower().str.replace(" ","_")
geometrical_df.columns=geometrical_df.columns.str.lower().str.replace(" ","_")
experimental_df=experimental_df.drop(columns=["blade__name"])
experimental_df.info()
geometrical_df.info()
```

### Inferences
1. Reading all the files in the Python environment and appended all the three versions for experiment data
   and blade geometry data. Now, we have two datasets that contain all the information.ie experimental_df and geometrical_df
2. Changing the variable names in line with Python identifier naming convention.

```python
geometrical_df.rename(columns={'adimensional_chord_-_c/r':'c/R','adimensional_radius_-_r/r':'r/R'},inplace=True)
geometrical_df['R']=geometrical_df['propeller_diameter'] / 2
geometrical_df['radius_distribution']=geometrical_df['r/R'] * geometrical_df['R']
geometrical_df['chord_distribution']=geometrical_df['c/R'] * geometrical_df['R']
geometrical_df
```

### Inferences
1. The adimensional ratios c/R and r/R provided in the geometry dataset are originally scaled values.
2. To perform real aerodynamic calculations (blade area, solidity, thrust modeling), we must convert them into actual chord and radius values.
3. After computing radius R, multiplying the ratios by the radius gives the actual chord distribution and radius distribution along the blade.
4. These real values form the foundation for calculating blade surface area using numerical integration (np.trapz), which is later used to compute key aerodynamic features such as blade area, disc area, total blade area, and solidity.

```python
experimental_df['number_of_blades'].value_counts()
blade_areas = geometrical_df.groupby(
    ['propeller_brand','propeller_diameter','propeller_pitch'],
    as_index=False
).apply(
    lambda g: pd.Series({
        'blade_area':np.trapz(g['chord_distribution'],g['radius_distribution'])
    })
).reset_index(drop=True)
blade_areas['disc_area'] = np.pi * (blade_areas['propeller_diameter'] / 2) ** 2
blade_areas=blade_areas.merge(
    experimental_df[['propeller_brand','number_of_blades','propeller_diameter','propeller_pitch']],
    on=['propeller_brand','propeller_diameter','propeller_pitch'],
    how='left'
)
blade_areas
blade_areas['total_blade_area'] = blade_areas['blade_area'] * blade_areas['number_of_blades']
blade_areas['solidity'] = blade_areas['total_blade_area']/ blade_areas['disc_area']
blade_areas
```

### Inferences 
1. The entire code converts raw geometric shape data into key aerodynamic features:
blade_area, disc_area, total_blade_area, solidity
2. These features directly influence thrust, power consumption, and efficiency.
3. They are essential inputs for machine learning models predicting propeller performance.

```python
experimental_df= experimental_df.merge(
    blade_areas[['propeller_brand','propeller_diameter','propeller_pitch','blade_area','disc_area','number_of_blades','total_blade_area','solidity']],
    on=['propeller_brand','number_of_blades','propeller_diameter','propeller_pitch'],
    how='left'
)
experimental_df['number_of_blades'].value_counts()
df_modelB=experimental_df.copy()
experimental_df = experimental_df.dropna(subset=['blade_name','disc_area','total_blade_area','solidity'])
experimental_df.isnull().sum()
```

### Inferences 
1. We merging the geometry features (blade area, disc area, total blade area, solidity) into the experimental dataset.
2. The merge is performed on: propeller_brand blade_name propeller_diameter.
3. This ensures each propeller gets its correct geometric features.
4. As we know there are many propellers that does not have geometric features so we are getting null values after merging.
5. We drop the rows which have null values so that we work on the propeller which has geometric features.

### Inferences
1. Thrust Coefficient (Ct): As we can see the plot shows right-skewed distribution and most values between 0.08-0.10, indicating the dataset contains mostly moderate-performance propellers.
2. Power Coefficient (Cp): Most values fall in the 0.04-0.06 range, meaning the propeller in the dataset are generally energy efficient with the moderate power consumption.
3. Efficiency: Most values between 0.6-1.0 is very good for UAV propellers. Negative or extremely low values indicate bad operating condition.
4. RPM: Clear Multiple peaks around (3000,4000,5000,6000,7000). This reflects structured test cycle rather than random moter speeds.

```python
# Heatmap Correlation Matrix
plt.figure(figsize=(14,8))
sns.heatmap(experimental_df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.tight_layout()
plt.title("Correlation matrix")
plt.show()
```

## Inferences 
1. High thrust requires high power: Ct and Cp have very strong positive correlation (0.88).
2. Forward speed reduces thrust: Ct and advanced_ratio_input have strong negative correlation (-0.82).
3. Pitch increases power consumption: Higher pitch -> more load -> more power.
4. Larger diameter = Lower RPM: Large props spin slower -> more efficient.
5. Solidity is weakly correlated with performance variables: solidity influence performance non-linearly (<0.2)
6. Very high multicollinearity between geometric features: blade_area, disc_area, total_blade_area all correlate > 0.93
7. Efficiency has low direct corrleation: Efficiency  vs others (<0.3) and it depends on combined features

## Machine Learning

### Training Model A - Without Missing Value Imputation 

```python
# Training Model A without missing value imputation
df_modelA=experimental_df.copy()
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,root_mean_squared_error,r2_score
# Define features and labels
X=df_modelA[['propeller_diameter','propeller_pitch','advanced_ratio_input','blade_area','disc_area','total_blade_area','solidity']]
y_CT=df_modelA['thrust_coefficient_output']
y_CP=df_modelA['power_coefficient_output']
y_EF=df_modelA['efficiency_output']
#Train-test split
Xtrain, Xtest, ytrain_ct, ytest_ct= train_test_split(X,y_CT,test_size=0.2,random_state=42)

Xtrain, Xtest, ytrain_cp, ytest_cp= train_test_split(X,y_CP,test_size=0.2,random_state=42)

Xtrain, Xtest, ytrain_ef, ytest_ef= train_test_split(X,y_EF,test_size=0.2,random_state=42)

# DEFINE XGBOOST MODEL
def create_xgb_model():
    return XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=5,
        subsample=0.9,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )
# TRAIN THREE MODELS: CT, CP, EF

gb_ct=create_xgb_model()
gb_ct.fit(Xtrain,ytrain_ct)

gb_cp=create_xgb_model()
gb_cp.fit(Xtrain,ytrain_cp)

gb_ef=create_xgb_model()
gb_ef.fit(Xtrain,ytrain_ef)

# MAKE PREDICTIONS
pred_ct=gb_ct.predict(Xtest)
pred_cp=gb_cp.predict(Xtest)
pred_ef=gb_ef.predict(Xtest)

# Evaluation metric of Thrust Coefficient 
mae_A_CT=mean_absolute_error(ytest_ct,pred_ct)*100
rmse_A_CT=root_mean_squared_error(ytest_ct,pred_ct)*100
r2_A_CT=r2_score(ytest_ct,pred_ct)*100

# Evaluation metric of Power Coefficient 
mae_A_CP=mean_absolute_error(ytest_cp,pred_cp)*100
rmse_A_CP=root_mean_squared_error(ytest_cp,pred_cp)*100
r2_A_CP=r2_score(ytest_cp,pred_cp)*100

# Evaluation metric of Efficiency 
mae_A_EF=mean_absolute_error(ytest_ef,pred_ef)*100
rmse_A_EF=root_mean_squared_error(ytest_ef,pred_ef)*100
r2_A_EF=r2_score(ytest_ef,pred_ef)*100
```

### Training Model B - With Missing Value Imputation

```python
df_modelB.isnull().sum()
df_modelB.info()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
cols_to_impute=['blade_area','disc_area','total_blade_area','solidity']
df_modelB[cols_to_impute]=imputer.fit_transform(df_modelB[cols_to_impute])
```

```python
# FILTER ONLY 3 & 4 BLADE DATA
df_34 = df_modelB[df_modelB['number_of_blades'] != 2].copy()

print("3/4-blade dataset shape:", df_34.shape)
print(df_34['number_of_blades'].value_counts())


# DEFINE FEATURES AND TARGETS
features = [
    'propeller_diameter',
    'propeller_pitch',
    'advanced_ratio_input',
    'blade_area',
    'disc_area',
    'total_blade_area',
    'solidity'
]

X = df_34[features]

y_CT = df_34['thrust_coefficient_output']
y_CP = df_34['power_coefficient_output']
y_EF = df_34['efficiency_output']


# TRAIN-TEST SPLITS
Xtrain_CT, Xtest_CT, ytrain_CT, ytest_CT = train_test_split(
    X, y_CT, test_size=0.2, random_state=42
)
Xtrain_CP, Xtest_CP, ytrain_CP, ytest_CP = train_test_split(
    X, y_CP, test_size=0.2, random_state=42
)
Xtrain_EF, Xtest_EF, ytrain_EF, ytest_EF = train_test_split(
    X, y_EF, test_size=0.2, random_state=42
)


# DEFINE XGBOOST MODEL
def create_xgb_model():
    return XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=5,
        subsample=0.9,
        colsample_bytree=0.8,
        random_state=42,
        n_jobs=-1
    )


# TRAIN THREE MODELS: CT, CP, EF
xgb_CT_34 = create_xgb_model()
xgb_CT_34.fit(Xtrain_CT, ytrain_CT)

xgb_CP_34 = create_xgb_model()
xgb_CP_34.fit(Xtrain_CP, ytrain_CP)

xgb_EF_34 = create_xgb_model()
xgb_EF_34.fit(Xtrain_EF, ytrain_EF)


# MAKE PREDICTIONS
# ---------------------------------------------------------
pred_CT = xgb_CT_34.predict(Xtest_CT)
pred_CP = xgb_CP_34.predict(Xtest_CP)
pred_EF = xgb_EF_34.predict(Xtest_EF)


# Evaluation metric of Thrust Coefficient 
mae_B_CT=mean_absolute_error(ytest_CT, pred_CT)*100
rmse_B_CT=root_mean_squared_error(ytest_CT, pred_CT)*100
r2_B_CT=r2_score(ytest_CT, pred_CT)*100

# Evaluation metric of Power Coefficient 
mae_B_CP=mean_absolute_error(ytest_CP, pred_CP)*100
rmse_B_CP=root_mean_squared_error(ytest_CP, pred_CP)*100
r2_B_CP=r2_score(ytest_CP, pred_CP)*100

# Evaluation metric of Efficiency 
mae_B_EF=mean_absolute_error(ytest_EF, pred_EF)*100
rmse_B_EF=root_mean_squared_error(ytest_EF, pred_EF)*100
r2_B_EF=r2_score(ytest_EF, pred_EF)*100
```

### Training Model C - Without Solidity Feature

```python
df_modelC = df_modelB.drop(columns=['solidity']).copy()
df_modelC['number_of_blades'].value_counts()
```

```python
#Define features and labels
X=df_modelC[['propeller_diameter','propeller_pitch','advanced_ratio_input','blade_area','disc_area','total_blade_area']]
y_CT=df_modelC['thrust_coefficient_output']
y_CP=df_modelC['power_coefficient_output']
y_EF=df_modelC['efficiency_output']
#Train-test split
Xtrain, Xtest, ytrain_ct, ytest_ct= train_test_split(X,y_CT,test_size=0.2,random_state=42)
#Train Model B for Thrust Coefficient 
gb_ct_C=GradientBoostingRegressor()
gb_ct_C.fit(Xtrain,ytrain_ct)
# Predict Thrust Coefficient 
pred_ct_C=gb_ct_C.predict(Xtest)
# Metrics
mae_C_CT=mean_absolute_error(ytest_ct,pred_ct_C)*100
rmse_C_CT=root_mean_squared_error(ytest_ct,pred_ct_C)*100
r2_C_CT=r2_score(ytest_ct,pred_ct_C)*100
#Train-test split
Xtrain, Xtest, ytrain_cp, ytest_cp= train_test_split(X,y_CP,test_size=0.2,random_state=42)
#Train Model B for Power Coefficient 
gb_cp_C=GradientBoostingRegressor()
gb_cp_C.fit(Xtrain,ytrain_cp)
# Predict Thrust Coefficient 
pred_cp_C=gb_cp_C.predict(Xtest)
# Metrics
mae_C_CP=mean_absolute_error(ytest_cp,pred_cp_C)*100
rmse_C_CP=root_mean_squared_error(ytest_cp,pred_cp_C)*100
r2_C_CP=r2_score(ytest_cp,pred_cp_C)*100
#Train-test split
Xtrain, Xtest, ytrain_ef, ytest_ef= train_test_split(X,y_EF,test_size=0.2,random_state=42)
#Train Model B for Efficiency 
gb_ef_C=GradientBoostingRegressor()
gb_ef_C.fit(Xtrain,ytrain_ef)
# Predict Thrust Coefficient 
pred_ef_C=gb_ef_C.predict(Xtest)
# Metrics
mae_C_EF=mean_absolute_error(ytest_ef,pred_ef_C)*100
rmse_C_EF=root_mean_squared_error(ytest_ef,pred_ef_C)*100
r2_C_EF=r2_score(ytest_ef,pred_ef_C)*100
```

```python
comparison_table = pd.DataFrame({
    'Model': ['Model A', 'Model B', 'Model C'],

    'MAE_CT':  [mae_A_CT, mae_B_CT, mae_C_CT],
    'RMSE_CT': [rmse_A_CT, rmse_B_CT, rmse_C_CT],
    'R2_CT':   [r2_A_CT, r2_B_CT, r2_C_CT],

    'MAE_CP':  [mae_A_CP, mae_B_CP, mae_C_CP],
    'RMSE_CP': [rmse_A_CP, rmse_B_CP, rmse_C_CP],
    'R2_CP':   [r2_A_CP, r2_B_CP, r2_C_CP],

    'MAE_EF':  [mae_A_EF, mae_B_EF, mae_C_EF],
    'RMSE_EF': [rmse_A_EF, rmse_B_EF, rmse_C_EF],
    'R2_EF':   [r2_A_EF, r2_B_EF, r2_C_EF],
})

comparison_table
```
